```python
import pandas as pd
import json
df1=pd.read_excel("ACOE_Validations_Domain_15032021.xlsx", sheet_name="To Extract")
```

```python
def fun(x):
    try:
        k=json.loads(x)
        k1=pd.DataFrame(k)
        k2=k1[['intent','confidence', 'domain']].reset_index().melt(id_vars='index')
        k2=k2.sort_values(by='index')
        k2['variable']=k2['variable']+'_'+k2['index'].astype(str)

        k3=pd.Series(data=k2['value'])
        k3.index=k2['variable']
        return k3
    except:
        pass
```

```python
df2=df1[~df1['Remarks'].isnull()]

df3=df2.join(df2['Remarks'].apply(lambda x: fun(x)))

df4=pd.concat([df3,df1[df1['Remarks'].isnull()]])
df4.fillna('None',inplace=True)

df4.to_csv('ACOEvalidation.csv',index=None)
```
